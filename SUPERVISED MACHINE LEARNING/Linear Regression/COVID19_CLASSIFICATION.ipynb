```python
# case in pakistan
pakistan_case=covid[covid["location"]=="Pakistan"]
pakistan_case.head()
```

```python
pakistan_case.tail()
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
#Total cases per day
sns.set(rc={'figure.figsize':(20,10)})
sns.lineplot(x="date",y="total_cases",data=pakistan_case)
plt.show()
```

```python
#Making a dataframe for last 5 days
pakistan_last_5_days=pakistan_case.tail()
#Total cases in last 5 days
sns.set(rc={'figure.figsize':(3,3)})
sns.lineplot(x="date",y="total_cases",data=pakistan_last_5_days)
plt.show()
```

```python
#Total tests per day
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date",y="total_tests",data=pakistan_case)
plt.show()
```

```python
#Total tests in last 5 days
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date",y="total_tests",data=pakistan_last_5_days)
plt.show()
```

```python
#china Case
china_case=covid[covid["location"]=="China"]
china_case.head()
```

```python
china_case.head(12)
```

```python
#Making a dataframe for china for last 5 days
china_last_5_days=china_case.tail()
#Total cases in last 5 days
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date",y="total_cases",data=china_last_5_days)
plt.show()
```

```python
#Understanding cases of pakistan, China and Japan
pakistan_japan_china=covid[(covid["location"] =="Pakistan") | (covid["location"] =="China") | (covid["location"]=="Japan")]
#Plotting growth of cases across China, India and Japan
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="location",y="total_cases",data=pakistan_japan_china,hue="date")
plt.show()
```

```python
#Understanding cases of germany and spain
germany_spain=covid[(covid["location"] =="Germany") | (covid["location"] =="Spain")]
#Plotting growth of cases across Germany and Spain
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="location",y="total_cases",data=germany_spain,hue="date")
plt.show()
```

```python
#Getting latest data
last_day_cases=covid[covid["date"]=="2020-05-24"]
last_day_cases
```

```python
#Sorting data w.r.t total_cases
max_cases_country=last_day_cases.sort_values(by="total_cases",ascending=False)
max_cases_country
```

```python
#Top 5 countries with maximum cases
max_cases_country[1:6]
```

```python
#Making bar-plot for countries with top cases
sns.barplot(x="location",y="total_cases",data=max_cases_country[1:6],hue="location")
plt.show()
```
